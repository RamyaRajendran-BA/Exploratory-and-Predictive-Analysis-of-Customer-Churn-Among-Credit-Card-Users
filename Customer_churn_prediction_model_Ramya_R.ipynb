# **Introduction**



*   In the dynamic and highly competitive credit card industry, customer retention is critical for sustaining long-term profitability. While acquiring new customers is essential, retaining existing ones is far more cost-effective and can significantly contribute to a company’s growth and success. However, despite best efforts, many credit card users discontinue their accounts, resulting in customer churn, a complex phenomenon that can be driven by a variety of factors, such as dissatisfaction with service, lack of engagement, changing financial needs, or more appealing offers from competing institutions. Consequently, understanding and mitigating churn has become an urgent priority for businesses seeking to maintain a stable customer base and optimize their financial performance.

*   As the credit card market continues to evolve, driven by technological advancements and shifting consumer expectations, businesses must adapt by employing sophisticated methods to predict and address churn. The ability to accurately forecast which customers are likely to leave offers a strategic advantage, enabling companies to proactively intervene with targeted retention efforts. These strategies can range from personalized offers and enhanced customer service to tailored loyalty programs that increase customer satisfaction and engagement.






* This study focuses on the **Exploratory and Predictive Analysis of Customer Churn Among Credit Card Users**, using data-driven techniques to identify underlying factors contributing to churn and develop predictive models that can anticipate customer behavior. By analyzing customer demographics, transaction history, usage patterns, and service interactions, this research aims to uncover key insights that businesses can leverage to improve retention. Predictive analytics, in particular, offers the potential to not only pinpoint at-risk customers but also to customize retention efforts in a way that resonates with individual preferences, enhancing the overall customer experience.


# **🌟How We Can Calculate Customer Churn Rate ?**


Customer churn rate is the number of customers you have lost divided by the total number of customers. To have an estimate you can segment your customers on the basis of the frequency of their purchase.



**Customer Churn Rate = No. of Customers lost/Total no. of customers (Period) x 100**

````python
#PERCENTAGE OF THE MISSING VALUES - DataFrame.....
def missing_data(credit_churn):
    """
    Args:
      credit_churn: DataFrame
    Returns:
    """
    total = credit_churn.isnull().sum().sort_values(ascending = False)
    Percentage = (credit_churn.isnull().sum()/credit_churn.isnull().count()*100).sort_values(ascending = False)
    return pd.concat([total, Percentage], axis=1, keys=['Total', 'Percentage'])
missing_data(credit_churn)
````

````python
# List of classifiers
classifiers = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'Bagging': BaggingClassifier(),
    'Extra Trees': ExtraTreesClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    'Decision Tree': DecisionTreeClassifier(),
    'SVM': SVC(probability=True),
    'KNN': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB()
}

# Dictionaries to store results
results_train = {}
results_test = {}

# K-Fold cross-validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

for name, clf in classifiers.items():
    # Cross-validation
    cv_results = cross_val_score(clf, X_train, y_train, cv=kfold, scoring='accuracy')
    results_train[name] = {
        'CrossVal_Score_Mean': cv_results.mean(),
        'CrossVal_Error': cv_results.std()
    }

    # Train the model
    clf.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = clf.predict(X_test)
    y_pred_proba = clf.predict_proba(X_test)[:, 1] if hasattr(clf, "predict_proba") else None

    # Evaluate the predictions
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba) if y_pred_proba is not None else 'N/A'
    clf_report = classification_report(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    # Store the evaluation metrics
    results_test[name] = {
        'Accuracy': accuracy,
        'F1_Score': f1,
        'ROC_AUC_Score': roc_auc,
        'Classification_Report': clf_report,
        'Confusion_Matrix': cm
    }

# Print the cross-validation results
for name, result in results_train.items():
    print(f"{name} (Training):")
    print(f"  CrossVal_Score_Mean: {result['CrossVal_Score_Mean']:.4f}")
    print(f"  CrossVal_Error: {result['CrossVal_Error']:.4f}")
    print()

# Print the test results
for name, result in results_test.items():
    print(f"{name} (Test):")
    print(f"  Accuracy: {result['Accuracy']:.4f}")
    print(f"  F1_Score: {result['F1_Score']:.4f}")
    print(f"  ROC_AUC_Score: {result['ROC_AUC_Score']}")
    print(f"  Classification_Report:\n{result['Classification_Report']}")
    print(f"  Confusion_Matrix:\n{result['Confusion_Matrix']}\n")
````

## 19. Retraining the Model with SMOTE to Improve True Positives

The previous confusion matrix showed that the model failed to identify any True Positives (churned customers). This is often due to an imbalanced dataset where the minority class (churned customers) is significantly underrepresented. To address this, we will use **SMOTE (Synthetic Minority Over-sampling Technique)** to generate synthetic samples for the minority class, thereby balancing the dataset. After balancing, we will retrain the Random Forest model using the best hyperparameters found previously and evaluate its performance.

### **19.1. Applying SMOTE to Balance the Training Data**

To address the severe class imbalance, SMOTE (Synthetic Minority Over-sampling Technique) was applied to the training dataset. This technique generates synthetic samples for the minority class ('Attrited Customer'), thereby increasing its representation and balancing the dataset. The test set remains untouched to ensure an unbiased evaluation of the model's performance on unseen data.
